```python
#Ejercicio 5

def matricesIguales(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        return False 
    else:
        for i in range(len(A)):
            for j in range(len(A[0])):
                if A[i][j] != B[i][j]:
                    return False
    return True

A = np.array([[4,2,1],[2,7,9],[0,5,22/3]])
L = np.array([[1., 0., 0.],[0.5, 1., 0.],[0., 5/6, 1.]])
U = np.array([[4., 2., 1.],[0., 6., 8.5],[0., 0., 0.25]])

print(matricesIguales(A,L@U))
```

```python
#Ejercicio 6

#esCuadrada y traspuesta fueron sacadas del anterior taller.

def esCuadrada(m) -> bool:
    lFilas = len(m)
    lColumnas = len(m[0])

    if lFilas != lColumnas:
        return False
    return True

def traspuesta(m):
    lF = len(m)
    lC = len(m[0])
    U = np.zeros((lC, lF))

    i = 0
    while i != lF:
        for j in range(lC):
            U[j][i] = m[i][j]
        i = i+1
    return U

def esSimetrica(m) -> bool:
    lF = len(m)
    lC = len(m[0])
    
    if not esCuadrada(m):
        return False
        
    n = traspuesta(m)

    for i in range(lF):
        for j in range(lC):
            if m[i][j] != n[i][j]:
                return False
    return True

A = np.array(np.random.rand(4,4))
esSimetrica(A)

print(esSimetrica(A.T@A))
print(esSimetrica(A.T@(A * 0.25)/0.25))
print(esSimetrica(A.T@(A * 0.2)/0.2))
```
